```python
import pandas as pd 

df = pd.read_csv("appartment.csv" , sep='\t')

df

```

```python
df['Price'] = df['Price'].str.replace(',', '').astype(float)
X = df[['Area', 'Rooms']]

Y = df['Price']

X
```

```python
from sklearn.tree import DecisionTreeRegressor

model = DecisionTreeRegressor()
model.fit(X, Y)


print("Im ready for training")
```

```python
def new_appertment (Area , Rooms) :

    data = [[Area, Rooms]]

    predection = model.predict(data)

    print( "The price of the appartment = ",predection[0])



    

```

```python
user_area = float(input("Enter the area : "))
user_rooms = int(input("Enter the numbers of rooms : "))
print("-"*30)
print("Deep thinking....")
print("-"*30)
new_appertment(user_area , user_rooms)
```
